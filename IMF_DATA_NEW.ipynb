```python
#lấy riêng từng chỉ tiêu cho từng cơ sở dữ liệu
import sdmx

IMF_DATA = sdmx.Client('IMF_DATA')

data_msg = IMF_DATA.data('WEO', key='VNM.NGDPD.A', params={'startPeriod': 2010, 'endPeriod': 2028})

df = sdmx.to_pandas(data_msg)

df
```

```python
#bỏ các cột thừa
df.index = df.index.get_level_values('TIME_PERIOD')
df
```

```python
#xuất dữ liệu đã tải ra một file excel
df.to_excel('imf1.xlsx')
```

```python
#tải nhiều chỉ tiêu và xuất ra các file excel
import sdmx

IMF_DATA = sdmx.Client('IMF_DATA')

for i in ['VNM.A_T.R_F.USD.A','VNM.A_T.R_FR2.USD.A']:
  data_msg = IMF_DATA.data('BOP', key=i, params={'startPeriod': 2010, 'endPeriod': 2028})
  df = sdmx.to_pandas(data_msg)
  df.index = df.index.get_level_values('TIME_PERIOD')
  df.to_excel(i+'.xlsx')
```

```python
#tải nhiều chỉ tiêu và gộp vào cùng 1 file excel
import sdmx
import pandas as pd

IMF_DATA = sdmx.Client('IMF_DATA')

df_full = pd.DataFrame()

for i in ['VNM.A_T.R_F.USD.A','VNM.A_T.R_FR2.USD.A']:
  data_msg = IMF_DATA.data('BOP', key=i, params={'startPeriod': 2010, 'endPeriod': 2028})
  df = sdmx.to_pandas(data_msg)
  df.index = df.index.get_level_values('TIME_PERIOD')
  df_full = pd.concat([df_full, df], axis=1)

df_full.columns = ['R_F','R_FR2']
df_full
```
